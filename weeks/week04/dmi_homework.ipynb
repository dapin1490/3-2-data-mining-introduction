- [\[Python NumPy\] ndarray 데이터 형태 지정 및 변경 (Data Types for ndarrays)](https://rfriend.tistory.com/285)

```python
import numpy as np

a = np.zeros(3, dtype=[('a', 'f2'), ('b', 'i2'), ('c', 'S2')])

# write under 8 lines -------------------------
a = np.zeros(10, dtype=[('a', 'f2'), ('b', 'i2'), ('c', 'S5')])
for i in range(0, 10):
    a[i][0] = (i + 1) ** 2
    a[i][1] = (i + 1) ** 5
    a[i][2] = 'a' * (i + 1)
# -------------------------

print(a)
```

```python
# 힌트
a1 = np.array([0, 1, 2, 3])
a2 = np.array([3, 2, 1, 0])
a3 = np.append(a1, a2)
print(a3)
```
